```python
# Promedio de tipo de show
promedio = data.groupby('rating')['show_id'].mean()
print(promedio)

# Gráfico de barras para visualizar
promedio.plot(kind='bar', color='purple')
plt.title('Promedio de tipo de show')
plt.ylabel('Promedio de shows')
plt.xlabel('Tipo de Show')
plt.show()
```

```python
# Cuenta de Año de Lanzamiento
year_lanzamiento = data.groupby('release_year')['show_id'].count()

# Graficar el año de lanzamiento
year_lanzamiento.plot(kind='pie', color='blue')
plt.title('Crecimiento Acumulado de las Ventas')
plt.ylabel('Ventas Acumuladas')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```
